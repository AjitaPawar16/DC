```python
data=data.drop(['S.No'],axis=1)
data.info() #deletes S.No column from the data
```

```python
from datetime import date
date.today().year
data['Car_Age']=date.today().year-data['Year']
data.head() #Creats a column 'Car_Age' which contains today's date minus the manufacturing year which gives our the age of the car
```

```python
data['Brand'] = data.Name.str.split().str.get(0)
#splits the string in column 'Name' of the data and creates 2 new columns as Brand and Model.
#The string is split with space as a separator and the first string is added in the column 'Brand'
```

```python
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
#2nd and 3rd string is added in the column 'Model'
```

```python
data[['Name','Brand','Model']] #selecting the 3 mentioned columns from the data
```

```python
print(data.Brand.unique()) #Listing all the unique entries in the column Brand
print(data.Brand.nunique()) #Number of unique entries in the column Brand
```

```python
searchfor = ['Isuzu' ,'ISUZU','Mini','Land']
data[data.Brand.str.contains('|'.join(searchfor))].head(5)
#Displaying first 5 entries of the dataset where the brand name is either one of - 'Isuzu' ,'ISUZU','Mini','Land'
```

```python
data["Brand"].replace({"ISUZU": "Isuzu", "Mini": "Mini Cooper","Land":"Land Rover"}, inplace=True)
#Replaces ISUZU with Isuzu and so on in the data set and not just in the output
```

```python
cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
#Separates the columns into Numerical and categorical types.
#cat_cols includes the columns that has categorical data. Selects the columns for which the data type 'object' (string)
#num_cols includes the columns that has numerical data. Selects the columns for which the data type is np.number (number)
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
sns.countplot(ax = axes[0, 0], x = 'Fuel_Type', data = data, color = 'blue',
              order = data['Fuel_Type'].value_counts().index);
sns.countplot(ax = axes[0, 1], x = 'Transmission', data = data, color = 'blue',
              order = data['Transmission'].value_counts().index);
sns.countplot(ax = axes[1, 0], x = 'Owner_Type', data = data, color = 'blue',
              order = data['Owner_Type'].value_counts().index);
sns.countplot(ax = axes[1, 1], x = 'Location', data = data, color = 'blue',
              order = data['Location'].value_counts().index);
sns.countplot(ax = axes[2, 0], x = 'Brand', data = data, color = 'blue',
              order = data['Brand'].head(20).value_counts().index);
sns.countplot(ax = axes[2, 1], x = 'Model', data = data, color = 'blue',
              order = data['Model'].head(20).value_counts().index);
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);
```
